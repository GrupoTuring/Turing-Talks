# Bibliotecas de Data Science 
## Matplotlib

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## Os principais tipos de plots
Dadas duas listas `x` e `y` de números reais, podemos plotar gráficos de **linha** e **dispersão** com eixos x e y.

````python
x = [1, 2, 7, 8, 10]
y = [2, 3, 2, 2, 4]
plt.plot(x, y) # Gráfico de linha
````

````python
plt.scatter(x, y) # Gráfico de dispersão
````

Para mostrar a frequência de várias categorias, podemos utilizar gráficos de **barra**, **setores** (pizza) e **histogramas**. Os gráficos de barra e setores recebem duas listas: uma com as categorias e outra com a frequência de cada categoria. Já o histograma recebe apenas uma lista `x`. Cada elementos de `x` representa uma ocorrência de uma categoria. Por exemplo, a lista `x = ['A', 'A', 'B', 'C']` indica que a categoria A ocorre duas vezes e que as categorias B e C ocorrem uma vez cada.

````python
x = ['A', 'B', 'C'] # Categorias
y = [3, 2, 1] # Número de ocorrências de cada categoria
plt.bar(x, y) # Gráfico de barras
````

````python
plt.pie(y, labels=x) # Gráfico de setores
````

````python
x = ['A', 'A', 'A', 'B', 'B', 'C']
plt.hist(x)
````

Pode parecer que o gráfico de barras e o histograma fazem a mesma coisa. No entanto, se trabalharmos com dados numéricos ao invés de categóricos (se os elementos da lista `x` forem números reais), perceberemos que o histograma agrupa valores próximos. Observe abaixo que o gráficos de barras (em cima) mostra cinco barras com $y = 1$ (as quatro barras para $x = 1, 1.01, 1.02, 1.03$ estão sobrepostas). Já o histograma agrupa esses valores próximos e mostra duas barras: uma com valor 4 em $x = 1$ e outra com valor 1 em $x = 2$.

````python
x = [1, 1.01, 1.02, 1.03, 2]
y = [1, 1, 1, 1, 1]

plt.bar(x, y)
plt.show()

plt.hist(x)
plt.show()

# Aqui usamos plt.show() para plotar dois gráficos separados. Caso contrário, o histograma e o gráfico de barras estariam superpostos (um em cima do outro).
````

## Customizando os gráficos
Os gráficos acima são bem simples, sem legendas, rótulos nos eixos x e y, ou cores, por exemplo. Aqui veremos as principais modificações que podemos fazer em nossos gráficos.

### Título, rótulos e legenda
Para colocar um título e rótulos nos eixos x e y, basta utilizar as funções `plt.title()`, `plt.xlabel()` e `plt.ylabel()`.

Para colocar uma legenda no gráfico, deve ser utilizada a função `plt.legend()`. Além disso, você deve passar um argumento `label` para as funções de plot com o nome de cada curva. Por exemplo:

````python
x = np.arange(1, 10, .1)
y = 5 * x**2
y2 = 2**x

plt.plot(x, y, label='5x^2')
plt.plot(x, y2, label='2^x')
plt.legend()
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Exemplo')
````

Obs: ao invés de passar o argumento `label`, você também pode passar as labels diretamente para a função legend: `plt.legend(['5x^2', '2^x'])`.

### Escala Logarítmica
Às vezes, precisamos plotar gráficos com eixos em escala logarítmica. No matplotlib, isso pode ser feito com as funções `plt.loglog()` (ambos os eixos em escala logarítmica), `plt.semilogx()` e `plt.semilogy()` (apenas um dos eixos em escala logarítmica). Por exemplo:

````python
plt.plot(x, y2)
plt.title('Escala linear em ambos os eixos')
plt.show()

plt.plot(x, y2)
plt.semilogy()
plt.title('Escala logarítmica no eixo y')
plt.show()
````

### Manipulando os eixos

Ao fazer um gráfico, pode ser importante controlar qual o valor mínimo e máximo de cada eixo. Por exemplo, se estivermos mostrando a frequência relativa de algumas categorias ($A = B = 30\%$, $C = 40\%$), é importante que os valores dos eixos comecem em 0 e terminem em 1. Para isso, usamos as funções `plt.xlim((left, right))` e `plt.ylim((top, bottom))`.

````python
plt.bar(['A', 'B', 'C'], [.3, .3, .4])
plt.show()

plt.bar(['A', 'B', 'C'], [.3, .3, .4])
plt.ylim((0, 1)) # Mostrar valores de y entre 0 e 1
plt.show()
````

Para mostrar gridlines nos gráficos, é usada a função `plt.grid()`. Podemos chamar `plt.grid(axis)` para ativar as gridlines no eixo especificado. Podemos passar `'x'`, `'y'` e `'both'` como valores de `axis`.

````python
plt.grid(axis='x')
plt.plot(x, y)
````

## Annotate e text
Para marcar algum ponto importante no gráfico, podemos usar a função `plt.annotate(text, (x, y), (xtext, ytext))`, onde (x, y) é o ponto a ser destacado e (xtext, ytext) é a localização do texto no gráfico.

````python
plt.plot(x, y)
plt.annotate('Algum texto útil', (5, 125), xytext=(3, 200), arrowprops={'arrowstyle': '->'})
# O argumento arrowprops faz com que seja desenhada uma seta do texto ao ponto
````

Se quisermos colocar algum texto no gráfico, mas esse texto não estiver relacionado a algum ponto específico, podemos usar a função `plt.text(x, y, text)`.

````python
plt.plot(x, y)
plt.text(2, 400, 'Algum comentário pertinente')
````

### Estilo das linhas e pontos
As características específicas de cada gráfico (de linhas, dispersão, etc.) são definidas quando se chama as funções de plot.

No caso da função `plt.plot()`, podemos passar um parâmetro `c` com a cor da linha e `linestyle` com o estilo da linha (`'-'` = normal, `'--'` = tracejado, `'-.'` = traço ponto, `':'` = pontilhado).

````python
plt.plot(x, y, c='red', linestyle='--')
````

A função `plt.scatter()` aceita um parâmetro `s` (um número) com o tamanho dos pontos, `marker` com o formato dos pontos e `c` com a cor dos pontos.

`marker` pode ser, por exemplo, `'x'` para indicar que os pontos devem ser desenhados em formato de x (veja mais tipos de pontos [aqui](https://matplotlib.org/api/markers_api.html#module-matplotlib.markers)).

`c` pode ser uma cor, uma lista de cores (uma para cada ponto) ou uma lista de números (um para cada ponto). Se for uma lista de números, as cores variarão conforme esses números, em uma [escala de cores](https://matplotlib.org/tutorials/colors/colormaps.html#classes-of-colormaps) determinada pelo parâmetro `cmap`.

````python
plt.scatter([1, 2, 3], [2, 3, 1], s=400, c='red')
plt.show()
````

````python
plt.scatter([1, 2, 3], [2, 3, 1], c=['red', 'green', 'blue'], marker='x')
plt.show()
````

````python
xx = np.array([1, 1, 1.5, 2, 2, 3])
yy = np.array([2, 3, 1,   2, 3, 3])
plt.scatter(xx, yy, c=xx+yy, cmap='winter')  # A cor representa a quantidade xx+yy
plt.show()
````

Quando usamos cores, podemos usar a função `plt.colorbar()` para criar uma legenda para as cores. No gráfico abaixo, usamos cores para indicar a diferença entre y e y2.

````python
z = y - y2
plt.plot(x, y2, c='red')
plt.scatter(x, y, c=z)
plt.text(x[-1], y[-1]+50, 'y')
plt.text(x[-1]-.6, y2[-1], 'y2')
plt.colorbar()
````

